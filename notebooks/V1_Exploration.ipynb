```python
# Kurze explorations der Dateien und deren Form


import pyarrow.dataset as ds
import pyarrow.parquet as pq
import os
import pandas as pd
import geopandas as gpd
from shapely import wkb
import fastparquet  # noqa

base_path = '/content/data'

unterordner = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]

for ordner in sorted(unterordner):
    ordner_pfad = os.path.join(base_path, ordner)

    print(f"\n{'='*70}")
    print(f"DATENSATZ: {ordner.upper()}")
    print(f"{'='*70}")

    try:
        # ── geo_information: WKB Hex -> lat/lon ──────────────────────────────
        if ordner == 'geo_information':
            table = pq.read_table(ordner_pfad)
            df = table.to_pandas()
            geom = df['location'].apply(lambda x: wkb.loads(bytes.fromhex(x)))
            df['latitude']  = geom.apply(lambda g: round(g.y, 6))
            df['longitude'] = geom.apply(lambda g: round(g.x, 6))
            df = df.drop(columns=['location'])

            print("Spalten und Datentypen:")
            for col, dtype in zip(df.columns, df.dtypes):
                print(f"- {col}: {dtype}")
            print("\nBeispielzeilen (erste 3):")
            display(df.head(3))

        # ── osm_landuse: WKB Bytes -> Polygon + Fläche ───────────────────────
        elif ordner == 'osm_landuse':
            df_raw = pd.read_parquet(ordner_pfad, engine='fastparquet')
            df_raw['area'] = df_raw['area'].apply(lambda x: wkb.loads(x))
            gdf = gpd.GeoDataFrame(df_raw, geometry='area', crs='EPSG:4326')
            gdf = gdf.to_crs('EPSG:3857')
            gdf['area_m2']  = gdf['area'].area.round(2)
            gdf['area_km2'] = (gdf['area_m2'] / 1_000_000).round(6)
            gdf = gdf.to_crs('EPSG:4326')
            gdf['area'] = gdf['area'].apply(lambda g: g.wkt)

            print("Spalten und Datentypen:")
            for col, dtype in zip(gdf.columns, gdf.dtypes):
                print(f"- {col}: {dtype}")
            print("\nBeispielzeilen (erste 3):")
            display(gdf[['id', 'city', 'landuse', 'area_m2', 'area_km2', 'area']].head(3))

        # ── alle anderen Datensätze: Standard PyArrow ─────────────────────────
        else:
            dataset = ds.dataset(ordner_pfad, format="parquet")
            print("Spalten und PyArrow-Datentypen:")
            for field in dataset.schema:
                print(f"- {field.name}: {field.type}")
            df_head = dataset.head(3).to_pandas()
            print("\nBeispielzeilen (erste 3):")
            display(df_head)

    except Exception as e:
        print(f"Fehler bei '{ordner}': {e}")
```

```python
# City-Exploration

# Pro Netzwerk
# Wie viele Demand-Zeilen?
# Wie viele unique Stations?: mit BIKE-XXX usw.
# Wie viel AVG-Daily Demand?
# ...

import pandas as pd
import numpy as np
from pathlib import Path


DEMAND_FOLDER = '/content/data/demand/'
GEO_FILE = '/content/data/geo_information/geo_information.parquet'
OUTPUT_CSV = '/content/drive/MyDrive/BA_Colab/City_Profiles_V5.csv'

# ==============================================================
# 1. DATEN LADEN
# ==============================================================
print("Lade Demand-Daten...")
demand_files = sorted(Path(DEMAND_FOLDER).rglob('*.parquet'))
print(f"  → {len(demand_files)} Parquet-Dateien gefunden")
df = pd.read_parquet([str(p) for p in demand_files])
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)
print(f"  → {len(df):,} Demand-Zeilen geladen")

print("Lade Geo-Daten...")
df_geo = (
    pd.read_parquet(GEO_FILE)
    .drop_duplicates(subset=['location_id'])
    [['location_id', 'city_name', 'federal_state_name']]
)

# Geo-Info an Demand joinen (nur zur Anreicherung)
df = df.merge(df_geo, on='location_id', how='left')

# ==============================================================
# 2. CITY PROFILING – gruppiert nach NETWORK_NAME
# ==============================================================
print("\nErstelle City Profiles nach network_name...")

profiles = []
for net_name, g in df.groupby('network_name'):

    # Sortieren nach Station und Zeit für sinnvolle Zeitdifferenzen
    g = g.sort_values(['station_id', 'timestamp'])

    # Zeitliche Abstände zwischen den Einträgen PRO STATION berechnen
    # Berechnet die Differenz zur vorherigen Zeile, gruppiert nach Station
    time_diffs = g.groupby('station_id')['timestamp'].diff()

    # Das häufigste Zeitintervall (Modus) in Minuten berechnen
    if time_diffs.notna().any():
        mode_diff = time_diffs.mode()
        if not mode_diff.empty:
            # Konvertiert timedelta zu lesbarem String (z.B. "0 days 01:00:00" -> "1.0 Stunden")
            mode_seconds = mode_diff.iloc[0].total_seconds()
            if mode_seconds == 3600:
                interval_str = "1 Stunde"
            elif mode_seconds == 86400:
                interval_str = "1 Tag"
            elif mode_seconds == 900:
                interval_str = "15 Minuten"
            else:
                # Fallback für krumme Werte: Umrechnung in Minuten
                interval_str = f"{int(mode_seconds / 60)} Min"
        else:
            interval_str = "Unbekannt"
    else:
        interval_str = "Zu wenig Daten"

    ts_min = g['timestamp'].min()
    ts_max = g['timestamp'].max()
    days_covered = (ts_max - ts_min).days + 1
    if days_covered < 1:
        days_covered = 1

    n_raw_rows = len(g)
    total_lends = int(g['n_lends'].sum())
    total_returns = int(g['n_returns'].sum())
    total_demand = total_lends + total_returns

    n_stations = g['station_name_id'].nunique() # Korrigiert auf echte Stations-IDs
    n_locations = g['location_id'].nunique()

    # Tägliche Nachfrage
    daily = g.set_index('timestamp').resample('D')[['n_lends', 'n_returns']].sum()
    daily['total'] = daily['n_lends'] + daily['n_returns']
    days_with_activity = int((daily['total'] > 0).sum())

    # Monatliche Saisonalität (CV)
    monthly = g.set_index('timestamp').resample('ME')['n_lends'].sum()
    monthly = monthly[monthly > 0]
    seasonality_cv = round(monthly.std() / monthly.mean(), 3) if len(monthly) > 1 and monthly.mean() > 0 else np.nan

    state = g['federal_state_name'].mode()
    state = state.iloc[0] if len(state) > 0 else 'Unknown'
    main_city = g['city_name'].mode()
    main_city = main_city.iloc[0] if len(main_city) > 0 else 'Unknown'

    profiles.append({
        'network_name': net_name,
        'main_city': main_city,
        'federal_state': state,
        'most_frequent_interval': interval_str, # <--- NEU
        'n_stations': n_stations,
        'n_locations': n_locations,
        'start_date': ts_min.strftime('%Y-%m-%d'),
        'end_date': ts_max.strftime('%Y-%m-%d'),
        'days_covered': days_covered,
        'days_with_activity': days_with_activity,
        'data_completeness': round(days_with_activity / days_covered, 3),
        'n_raw_rows': n_raw_rows,
        'total_lends': total_lends,
        'total_returns': total_returns,
        'total_demand': total_demand,
        'avg_daily_demand': round(total_demand / days_covered, 1),
        'avg_daily_demand_per_station': round(total_demand / days_covered / max(n_stations, 1), 2),
        'seasonality_cv': seasonality_cv,
    })

df_profiles = (
    pd.DataFrame(profiles)
    .sort_values('total_demand', ascending=False)
    .reset_index(drop=True)
)

# ==============================================================
# 4. OUTPUT
# ==============================================================
df_profiles.to_csv(OUTPUT_CSV, index=False)

print(f"\n{'='*60}")
print(f"ERGEBNIS: {len(df_profiles)} Netzwerke profiliert")
print(f"{'='*60}\n")

# Top 15 - Jetzt inkl. Intervall
print("Top 15 Netzwerke nach Gesamtnachfrage:")
print(df_profiles[[
    'network_name', 'main_city', 'n_stations',
    'n_raw_rows', 'most_frequent_interval', 'avg_daily_demand'
]].head(15).to_string(index=False))
```

```python
# Parquets to CSV (ohne Demand)
import pyarrow.dataset as ds
import pyarrow.parquet as pq
import pandas as pd
import geopandas as gpd
from shapely import wkb
from pathlib import Path

OUTPUT_DIR = Path('/content/data/csv_exports')
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)

DATA_BASE       = Path('/content/data')
excluded_folders = ['demand', 'csv_exports']

print("Exportiere Tabellen als CSV (außer DEMAND und csv_exports)...\n")

for folder in sorted(DATA_BASE.iterdir()):
    if not folder.is_dir():
        continue
    if folder.name.lower() in excluded_folders:
        continue

    print(f"→ Verarbeite: {folder.name}")
    out_path = (OUTPUT_DIR / folder.name).with_suffix('.csv')

    try:
        # ── geo_information: WKB Hex -> lat/lon ──────────────────────────────
        if folder.name == 'geo_information':
            df = pq.read_table(folder).to_pandas()
            geom = df['location'].apply(lambda x: wkb.loads(bytes.fromhex(x)))
            df['latitude']  = geom.apply(lambda g: round(g.y, 6))
            df['longitude'] = geom.apply(lambda g: round(g.x, 6))
            df = df.drop(columns=['location'])

        # ── osm_landuse: WKB Bytes -> Polygon + Fläche ───────────────────────
        elif folder.name == 'osm_landuse':
            df = pd.read_parquet(folder, engine='fastparquet')
            df['area'] = df['area'].apply(lambda x: wkb.loads(x))
            gdf = gpd.GeoDataFrame(df, geometry='area', crs='EPSG:4326')
            gdf = gdf.to_crs('EPSG:3857')
            gdf['area_m2']  = gdf['area'].area.round(2)
            gdf['area_km2'] = (gdf['area_m2'] / 1_000_000).round(6)
            gdf = gdf.to_crs('EPSG:4326')
            gdf['area'] = gdf['area'].apply(lambda g: g.wkt)
            df = pd.DataFrame(gdf[['id', 'city', 'landuse', 'area_m2', 'area_km2', 'area']])

        # ── alle anderen: Standard PyArrow ────────────────────────────────────
        else:
            df = ds.dataset(folder, format="parquet").to_table().to_pandas()

        df.to_csv(out_path, index=False, encoding='utf-8')
        print(f"   ✓ {len(df):,} Zeilen → {out_path}\n")

    except Exception as e:
        print(f"   ✗ FEHLER in {folder.name}: {e}\n")

print("Export abgeschlossen.")
```

```python
# Classification of BSS-System (stationbased, freefloatin, hybrid)
# Demand Zeilen werden Kategorisiert
# BIKE-XXX / All Rows. Je größer Quotient desto eher free floating
# Network_stats.csv hab ich runter geladen und ins Repo, aber nochmal die Thresholds angepasst per Nachbearbeitung

"""
=============================================================
  Nextbike Demand Dataset — Cleaning & Network Analysis
=============================================================
Schritte:
  1. Join DEMAND mit STATION_NAMES
  2. Recording-Einträge entfernen
  3. Pro Netzwerk: Anteil BIKE / virtuell / echte Stationen
  4. Exportiert bereinigte DataFrames + Übersichtstabelle
=============================================================
"""

import pandas as pd
import re

# ─── 0. DATEN LADEN ──────────────────────────────────────────────────────────
# Passe die Pfade auf deine lokale Umgebung an

demand       = pd.read_parquet("/content/data/demand/")        # oder .csv / .feather
station_names = pd.read_parquet("/content/data/station_names/station_names.parquet")

print(f"[INFO] DEMAND Zeilen (raw):          {len(demand):>10,}")
print(f"[INFO] STATION_NAMES Einträge:       {len(station_names):>10,}")

# ─── 1. JOIN: station_name_id → echter Name ──────────────────────────────────
# station_names hat Spalten: id (int32), name (object)
station_names = station_names.rename(columns={"id": "station_name_id", "name": "station_name"})

demand = demand.merge(station_names, on="station_name_id", how="left")

# Leere Namen auffüllen (falls kein Match)
demand["station_name"] = demand["station_name"].fillna("UNKNOWN")

print(f"[INFO] Nach Join — DEMAND Zeilen:    {len(demand):>10,}")
print(f"[INFO] Ohne Name-Match (UNKNOWN):    {(demand['station_name'] == 'UNKNOWN').sum():>10,}")

# ─── 2. STATION-NAME KLASSIFIKATION ──────────────────────────────────────────

def classify_station(name: str) -> str:
    """
    Kategorisiert einen Station-Namen in:
      - 'recording'  → Test/Dummy-Einträge    → ENTFERNEN
      - 'bike'       → Free-Floating Return   → SEPARIEREN
      - 'virtual'    → Virtuelle Station      → SEPARIEREN
      - 'only_nums'  → Nur Zahlen/Sonderzeichen (kein echter Name) → PRÜFEN
      - 'real'       → Echter Stationsname    → BEHALTEN
    """
    if not isinstance(name, str) or name.strip() == "":
        return "unknown"

    n = name.strip()

    if re.search(r'(?i)^recording[_\-\s]', n):
        return "recording"

    if re.match(r'(?i)^bike[-_]?\s*\d*', n):
        return "bike"

    if re.search(r'(?i)(virtuell|virtual)', n):
        return "virtual"

    # Nur Zahlen, Leerzeichen, Bindestriche, Unterstriche, Punkte — kein echter Name
    if re.fullmatch(r'[\d\s\-_\.#/]+', n):
        return "only_nums"

    return "real"

demand["station_type"] = demand["station_name"].apply(classify_station)

# Überblick über Verteilung
print("\n[INFO] Globale Verteilung station_type:")
print(demand["station_type"].value_counts().to_string())

# ─── 3. RECORDING-EINTRÄGE ENTFERNEN ─────────────────────────────────────────
recording_count = (demand["station_type"] == "recording").sum()
print(f"\n[CLEAN] Entferne {recording_count:,} Recording-Zeilen...")

demand_clean = demand[demand["station_type"] != "recording"].copy()
demand_recording = demand[demand["station_type"] == "recording"].copy()   # archiviert

print(f"[CLEAN] DEMAND nach Entfernung:      {len(demand_clean):>10,}")

# ─── 4. PRO NETZWERK: BIKE / VIRTUAL / REAL ──────────────────────────────────

def network_summary(df: pd.DataFrame) -> pd.DataFrame:
    """
    Erstellt eine Zusammenfassung pro Netzwerk:
    - Gesamtanzahl Demand-Zeilen
    - Anzahl und Anteil je station_type
    - Anzahl unique echter Stationen
    """
    groups = []

    for network, grp in df.groupby("network_name"):
        total_rows = len(grp)

        # Demand-Zeilen je Typ
        type_counts = grp["station_type"].value_counts()

        n_real_rows    = type_counts.get("real", 0)
        n_bike_rows    = type_counts.get("bike", 0)
        n_virtual_rows = type_counts.get("virtual", 0)
        n_nums_rows    = type_counts.get("only_nums", 0)
        n_unknown_rows = type_counts.get("unknown", 0)

        # Unique Stationen je Typ
        unique_real    = grp.loc[grp["station_type"] == "real",    "station_name_id"].nunique()
        unique_bike    = grp.loc[grp["station_type"] == "bike",    "station_name_id"].nunique()
        unique_virtual = grp.loc[grp["station_type"] == "virtual", "station_name_id"].nunique()

        # Anteil Free-Floating Returns (BIKE) an allen Zeilen → Systemtyp-Indikator
        bike_share = round(n_bike_rows / total_rows * 100, 2) if total_rows > 0 else 0

        groups.append({
            "network_name":      network,
            "total_rows":        total_rows,
            # --- Demand-Zeilen ---
            "rows_real":         n_real_rows,
            "rows_bike":         n_bike_rows,
            "rows_virtual":      n_virtual_rows,
            "rows_only_nums":    n_nums_rows,
            "rows_unknown":      n_unknown_rows,
            # --- Unique Stationen ---
            "unique_real_stations":    unique_real,
            "unique_bike_stations":    unique_bike,
            "unique_virtual_stations": unique_virtual,
            # --- Schlüsselkennzahl ---
            "bike_return_pct":   bike_share,   # > ~15% → wahrsch. Free-Floating/Hybrid
        })

    summary = pd.DataFrame(groups).sort_values("total_rows", ascending=False)
    return summary

print("\n[INFO] Erstelle Netzwerk-Zusammenfassung...")
network_stats = network_summary(demand_clean)

# ─── 5. SYSTEMTYP-KLASSIFIKATION PRO NETZWERK ────────────────────────────────
# Schwellenwert: bike_return_pct
#   < 5%  → stationbasiert (BSS klassisch)
#   5-15% → wahrscheinlich Hybrid
#   > 15% → Free-Floating-dominant oder Hybrid

THRESHOLD_FF   = 15.0   # % → Free-Floating
THRESHOLD_HYB  = 5.0    # % → Hybrid (zwischen 5 und 15)

def system_type(pct: float) -> str:
    if pct < THRESHOLD_HYB:
        return "station_based"
    elif pct < THRESHOLD_FF:
        return "hybrid"
    else:
        return "free_floating"

network_stats["system_type"] = network_stats["bike_return_pct"].apply(system_type)

print("\n[INFO] Systemtyp-Verteilung über Netzwerke:")
print(network_stats["system_type"].value_counts().to_string())

# ─── 6. SEPARIERUNG DER HAUPT-DATAFRAMES ─────────────────────────────────────

station_based_networks = network_stats.loc[
    network_stats["system_type"] == "station_based", "network_name"
].tolist()

demand_station_only = demand_clean[
    (demand_clean["network_name"].isin(station_based_networks)) &
    (demand_clean["station_type"] == "real")
].copy()

demand_bike_returns = demand_clean[demand_clean["station_type"] == "bike"].copy()
demand_virtual      = demand_clean[demand_clean["station_type"] == "virtual"].copy()

print(f"\n[RESULT] demand_station_only (Kern-Datensatz):  {len(demand_station_only):>10,} Zeilen")
print(f"[RESULT] demand_bike_returns (archiviert):       {len(demand_bike_returns):>10,} Zeilen")
print(f"[RESULT] demand_virtual      (archiviert):       {len(demand_virtual):>10,} Zeilen")
print(f"[RESULT] demand_recording    (archiviert):       {len(demand_recording):>10,} Zeilen")

# ─── 7. EXPORT ────────────────────────────────────────────────────────────────

network_stats.to_csv("network_stats.csv", index=False)
demand_station_only.to_parquet("demand_station_only.parquet", index=False)

# Archiv (für spätere Sensitivitätsanalysen)
demand_bike_returns.to_parquet("demand_bike_returns_archive.parquet", index=False)
demand_virtual.to_parquet("demand_virtual_archive.parquet", index=False)
demand_recording.to_parquet("demand_recording_archive.parquet", index=False)

print("\n[DONE] Alle Dateien exportiert.")
print("       → network_stats.csv             (Übersichtstabelle)")
print("       → demand_station_only.parquet   (Kern-Datensatz für Analyse)")
print("       → demand_*_archive.parquet      (Archiv, nicht gelöscht)")

# ─── 8. QUICK-PREVIEW: Top 20 Netzwerke ──────────────────────────────────────

print("\n" + "="*90)
print("TOP 20 NETZWERKE nach Demand-Zeilen")
print("="*90)
cols_preview = [
    "network_name", "system_type", "total_rows",
    "unique_real_stations", "bike_return_pct",
    "rows_real", "rows_bike", "rows_virtual"
]
print(network_stats[cols_preview].head(20).to_string(index=False))
```
