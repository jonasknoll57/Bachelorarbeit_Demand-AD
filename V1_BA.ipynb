```python
import pyarrow.dataset as ds
import os
import pandas as pd
import geopandas as gpd


# Hauptverzeichnis (Ihr erstellter symbolischer Link)
base_path = '/content/data'

# Alle Unterordner im Verzeichnis auflisten
unterordner = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]

for ordner in sorted(unterordner):
    ordner_pfad = os.path.join(base_path, ordner)

    try:
        # 1. PyArrow Dataset initialisieren (Lazy Loading)
        dataset = ds.dataset(ordner_pfad, format="parquet")

        print(f"\n{'='*70}")
        print(f"DATENSATZ: {ordner.upper()}")
        print(f"{'='*70}")

        # 2. Schema direkt aus den Metadaten auslesen
        print("Spalten und PyArrow-Datentypen:")
        for field in dataset.schema:
            print(f"- {field.name}: {field.type}")

        # 3. KORREKTUR: Die ersten 3 Zeilen mit head() extrahieren
        # head(3) holt genau 3 Zeilen als Arrow-Tabelle, die wir dann in Pandas umwandeln
        head_table = dataset.head(3)
        df_head = head_table.to_pandas()

        print("\nBeispielzeilen (erste 3):")
        display(df_head)

    except Exception as e:
        print(f"\nOrdner '{ordner}' konnte nicht verarbeitet werden.")
        print(f"Fehlergrund: {e}")
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

def create_city_profiles_fast(demand_folder, geo_file_path, output_csv='City_Profiles_Results.csv'):
    print(f"Lade Geo-Daten aus {geo_file_path}...")
    df_geo = pd.read_parquet(geo_file_path)

    # 1. Alle Dateien auf einen Schlag über PyArrow in C++ laden (extrem schnell)
    print(f"Lade ALLE Parquet-Dateien aus {demand_folder} auf einmal...")

    # Sucht rekursiv in allen Unterordnern nach .parquet Dateien
    demand_files = [str(p) for p in Path(demand_folder).rglob('*.parquet')]

    if not demand_files:
        print("FEHLER: Keine Parquet-Dateien gefunden! Bitte Pfad überprüfen.")
        return None

    print(f"Gefunden: {len(demand_files)} Dateien. Lese in den Arbeitsspeicher...")
    df_demand = pd.read_parquet(demand_files)

    print("Verarbeite Zeitstempel und berechne Metriken...")
    df_demand['timestamp'] = pd.to_datetime(df_demand['timestamp'])
    df_demand['month'] = df_demand['timestamp'].dt.month
    df_demand['total_demand'] = df_demand['n_lends'] + df_demand['n_returns']

    # 2. Zusammenführen mit Geo-Daten
    df_merged = pd.merge(df_demand, df_geo[['location_id', 'city_name', 'federal_state_name']],
                         on='location_id', how='left')

    print("Aggregiere Profile pro Stadt...")
    profiles = []

    # Gruppieren und Berechnen
    for city_name, city_df in df_merged.groupby('city_name'):
        start_date = city_df['timestamp'].min()
        end_date = city_df['timestamp'].max()
        days_active = (end_date - start_date).days + 1
        if days_active == 0: days_active = 1

        n_stations = city_df['station_id'].nunique()
        total_lends = city_df['n_lends'].sum()
        total_returns = city_df['n_returns'].sum()
        avg_daily_activity = (total_lends + total_returns) / days_active

        # Saisonalität (Variationskoeffizient)
        monthly_demand = city_df.groupby('month')['total_demand'].sum()
        seasonality_score = monthly_demand.std() / monthly_demand.mean() if monthly_demand.mean() > 0 else 0

        profiles.append({
            'city_name': city_name,
            'location_id': city_df['location_id'].iloc[0],
            'state': city_df['federal_state_name'].iloc[0],
            'n_stations': n_stations,
            'start_date': start_date.strftime('%Y-%m-%d'),
            'end_date': end_date.strftime('%Y-%m-%d'),
            'days_active': days_active,
            'total_lends': total_lends,
            'total_returns': total_returns,
            'avg_daily_activity': round(avg_daily_activity, 2),
            'seasonality_score': round(seasonality_score, 3)
        })

    df_final = pd.DataFrame(profiles).sort_values('avg_daily_activity', ascending=False).reset_index(drop=True)
    df_final.to_csv(output_csv, index=False)

    print(f"\n==========================================")
    print(f"Fertig! Datei lokal gespeichert als: {output_csv}")
    print(f"==========================================\n")

    print("=== TOP 3 STÄDTE (Potenzielle Source Cities) ===")
    print(df_final.head(3)[['city_name', 'n_stations', 'avg_daily_activity']])

    print("\n=== BOTTOM 3 STÄDTE (Potenzielle Target Cities) ===")
    print(df_final.tail(3)[['city_name', 'n_stations', 'avg_daily_activity']])

    return df_final

# ==========================================
# AUSFÜHRUNG MIT NEUEN PFADEN
# ==========================================
DEMAND_FOLDER = '/content/data/demand/'           # <--- Angepasster Pfad
GEO_FILE = '/content/data/geo_information/geo_information.parquet' # <--- Angepasster Pfad
OUTPUT_PATH = '/content/drive/MyDrive/BA_Colab/City_Profiles_Results.csv'

# Ausführen der Funktion
df_results = create_city_profiles_fast(DEMAND_FOLDER, GEO_FILE, OUTPUT_PATH)
```
