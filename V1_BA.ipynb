```python
import pandas as pd
import numpy as np
from pathlib import Path

# ==============================================================
# PFADE ANPASSEN
# ==============================================================
DEMAND_FOLDER = '/content/data/demand/'
GEO_FILE = '/content/data/geo_information/geo_information.parquet'
OUTPUT_CSV = '/content/drive/MyDrive/BA_Colab/City_Profiles_V5.csv'

# ==============================================================
# 1. DATEN LADEN
# ==============================================================
print("Lade Demand-Daten...")
demand_files = sorted(Path(DEMAND_FOLDER).rglob('*.parquet'))
print(f"  → {len(demand_files)} Parquet-Dateien gefunden")
df = pd.read_parquet([str(p) for p in demand_files])
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)
print(f"  → {len(df):,} Demand-Zeilen geladen")

print("Lade Geo-Daten...")
df_geo = (
    pd.read_parquet(GEO_FILE)
    .drop_duplicates(subset=['location_id'])
    [['location_id', 'city_name', 'federal_state_name']]
)

# Geo-Info an Demand joinen (nur zur Anreicherung)
df = df.merge(df_geo, on='location_id', how='left')

# ==============================================================
# 2. CITY PROFILING – gruppiert nach NETWORK_NAME
# ==============================================================
print("\nErstelle City Profiles nach network_name...")

profiles = []
for net_name, g in df.groupby('network_name'):

    # Sortieren nach Station und Zeit für sinnvolle Zeitdifferenzen
    g = g.sort_values(['station_id', 'timestamp'])

    # Zeitliche Abstände zwischen den Einträgen PRO STATION berechnen
    # Berechnet die Differenz zur vorherigen Zeile, gruppiert nach Station
    time_diffs = g.groupby('station_id')['timestamp'].diff()

    # Das häufigste Zeitintervall (Modus) in Minuten berechnen
    if time_diffs.notna().any():
        mode_diff = time_diffs.mode()
        if not mode_diff.empty:
            # Konvertiert timedelta zu lesbarem String (z.B. "0 days 01:00:00" -> "1.0 Stunden")
            mode_seconds = mode_diff.iloc[0].total_seconds()
            if mode_seconds == 3600:
                interval_str = "1 Stunde"
            elif mode_seconds == 86400:
                interval_str = "1 Tag"
            elif mode_seconds == 900:
                interval_str = "15 Minuten"
            else:
                # Fallback für krumme Werte: Umrechnung in Minuten
                interval_str = f"{int(mode_seconds / 60)} Min"
        else:
            interval_str = "Unbekannt"
    else:
        interval_str = "Zu wenig Daten"

    ts_min = g['timestamp'].min()
    ts_max = g['timestamp'].max()
    days_covered = (ts_max - ts_min).days + 1
    if days_covered < 1:
        days_covered = 1

    n_raw_rows = len(g)
    total_lends = int(g['n_lends'].sum())
    total_returns = int(g['n_returns'].sum())
    total_demand = total_lends + total_returns

    n_stations = g['station_name_id'].nunique() # Korrigiert auf echte Stations-IDs
    n_locations = g['location_id'].nunique()

    # Tägliche Nachfrage
    daily = g.set_index('timestamp').resample('D')[['n_lends', 'n_returns']].sum()
    daily['total'] = daily['n_lends'] + daily['n_returns']
    days_with_activity = int((daily['total'] > 0).sum())

    # Monatliche Saisonalität (CV)
    monthly = g.set_index('timestamp').resample('ME')['n_lends'].sum()
    monthly = monthly[monthly > 0]
    seasonality_cv = round(monthly.std() / monthly.mean(), 3) if len(monthly) > 1 and monthly.mean() > 0 else np.nan

    state = g['federal_state_name'].mode()
    state = state.iloc[0] if len(state) > 0 else 'Unknown'
    main_city = g['city_name'].mode()
    main_city = main_city.iloc[0] if len(main_city) > 0 else 'Unknown'

    profiles.append({
        'network_name': net_name,
        'main_city': main_city,
        'federal_state': state,
        'most_frequent_interval': interval_str, # <--- NEU
        'n_stations': n_stations,
        'n_locations': n_locations,
        'start_date': ts_min.strftime('%Y-%m-%d'),
        'end_date': ts_max.strftime('%Y-%m-%d'),
        'days_covered': days_covered,
        'days_with_activity': days_with_activity,
        'data_completeness': round(days_with_activity / days_covered, 3),
        'n_raw_rows': n_raw_rows,
        'total_lends': total_lends,
        'total_returns': total_returns,
        'total_demand': total_demand,
        'avg_daily_demand': round(total_demand / days_covered, 1),
        'avg_daily_demand_per_station': round(total_demand / days_covered / max(n_stations, 1), 2),
        'seasonality_cv': seasonality_cv,
    })

df_profiles = (
    pd.DataFrame(profiles)
    .sort_values('total_demand', ascending=False)
    .reset_index(drop=True)
)

# ==============================================================
# 4. OUTPUT
# ==============================================================
df_profiles.to_csv(OUTPUT_CSV, index=False)

print(f"\n{'='*60}")
print(f"ERGEBNIS: {len(df_profiles)} Netzwerke profiliert")
print(f"{'='*60}\n")

# Top 15 - Jetzt inkl. Intervall
print("Top 15 Netzwerke nach Gesamtnachfrage:")
print(df_profiles[[
    'network_name', 'main_city', 'n_stations',
    'n_raw_rows', 'most_frequent_interval', 'avg_daily_demand'
]].head(15).to_string(index=False))
```

```python
import pyarrow.dataset as ds
import pandas as pd
from pathlib import Path

# Pfad, in den alle CSVs geschrieben werden sollen
OUTPUT_DIR = Path('/content/data/csv_exports')
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)

# Basis für alle anderen Datensätze (ohne DEMAND und ohne csv_exports)
DATA_BASE = Path('/content/data')

# Auszuschließende Ordner (DEMAND und der CSV‑Export‑Ordner)
excluded_folders = ['DEMAND', 'csv_exports']  # klein/großschreibung wird unten ignoriert

print("Exportiere Tabellen als CSV (außer DEMAND und csv_exports)...")

for folder in DATA_BASE.iterdir():
    if not folder.is_dir():
        continue

    # Skip ausgeschlossene Ordner (case‑insensitive)
    if folder.name.lower() in [f.lower() for f in excluded_folders]:
        continue

    print(f"  → Verarbeite: {folder.name}")

    # Versuche, ein Parquet‑Dataset im Ordner zu laden
    try:
        dataset = ds.dataset(folder, format="parquet")
        df = dataset.to_table().to_pandas()

        # Ausgabedatei: z.B. /content/data/csv_exports/geo_information.csv
        out_path = (OUTPUT_DIR / folder.name).with_suffix('.csv')
        df.to_csv(out_path, index=False)
        print(f"      {len(df):,} Zeilen → {out_path}")

    except Exception as e:
        print(f"      FEHLER in {folder.name}: {e}")
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.metrics.pairwise import cosine_similarity

# ===================================================================
# 1. Pfade und Zielstädte (Netzwerke bzw. Stadt‑Mapping)
# ===================================================================

DEMAND_FOLDER = Path('/content/data/demand/')
GEO_FILE = '/content/data/geo_information/geo_information.parquet'
OUTPUT_CSV = '/content/drive/MyDrive/BA_Colab/week_profiles_city.csv'

# Liste der Städte, die du vergleichen willst (network_name oder via city_name)
# Stell dir vor, dass du z.B. über city_name oder ein mapping network_name → Stadt hast:
TARGET_CITIES = [
    'Berlin',
    'Köln',
    'Potsdam',
    'Marburg',
    'Kaiserslautern',
    'Ludwigshafen',
    'Neustadt an der Weinstraße'
]

# Falls du später wirklich nur bestimmte network_names nutzen willst,
# könntest du hier ein mapping definieren:
# NETWORK_BY_CITY = {
#     'Berlin': 'VRNnextbike_Berlin',
#     'Köln': 'VRNnextbike_Koeln',
#     ...
# }

print("Lade Demand-Daten...")
demand_files = sorted(DEMAND_FOLDER.rglob('*.parquet'))
print(f"  → {len(demand_files)} Parquet-Dateien gefunden")
df = pd.read_parquet([str(p) for p in demand_files])
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)

print("Lade Geo-Daten...")
df_geo = (
    pd.read_parquet(GEO_FILE)
    .drop_duplicates(subset=['location_id'])
    [['location_id', 'city_name']]
)

# Join von Demand mit Geo, um city_name zu bekommen (falls nötig)
df = df.merge(df_geo, on='location_id', how='left')


# ===================================================================
# 2. Filtere Prozessierung auf ausgewählte Netzwerke/Städte
# ===================================================================

# Wenn du wirklich nach network_name arbeitest, z.B.:
# TARGET_NETWORKS = ['network_name_berlin', 'network_name_koeln', ...]

# Alternativ: Du willst die Liste TARGET_CITIES oben als city_name‑Filter nutzen:
city_mask = df['city_name'].isin(TARGET_CITIES)
df_target = df[city_mask].copy()

print(f"  → {len(df_target):,} Zeilen nach Städtefilter")


# ===================================================================
# 3. Demand berechnen und auf stündliche Granularität resampeln
# ===================================================================

# Definiere total_demand pro Zeil
df_target = df_target.assign(total_demand=df_target['n_lends'] + df_target['n_returns'])

# Gruppiere nach Stadt (city_name) und berechne stündliche Summe
weekly_profiles = {}
for city in TARGET_CITIES:
    print(f"\nBearbeite Stadt: {city}")

    g = df_target[df_target['city_name'] == city].copy()

    # Sortierung nach Zeit und Location (falls multi‑location innerhalb Stadt)
    g = g.sort_values(['location_id', 'timestamp'])

    # Setze timestamp als Index und resample auf 1H (stündlich)
    g_hour = (
        g.set_index('timestamp')
        .groupby('location_id')  # optional: falls du stationen‑basiert bist
        [['total_demand']]
        .resample('1H')
        .sum()
        .reset_index()
    )

    # Nun aggregiere über alle location_id in dieser Stadt zur finalen stündlichen Serie
    g_city_hour = (
        g_hour
        .groupby('timestamp')['total_demand']
        .sum()
    )

    # 1. Wochentag und Stunde extrahieren (0=Mo, 1=Di, ..., 6=So)
    g_city_hour.index = g_city_hour.index.tz_localize(None)  # falls nötig
    g_city_hour = g_city_hour.reset_index()
    g_city_hour['weekday'] = g_city_hour['timestamp'].dt.dayofweek
    g_city_hour['hour'] = g_city_hour['timestamp'].dt.hour

    # 2. Gruppieren nach Wochentag (0–6) und Stunde (0–23)
    weekly = (
        g_city_hour
        .groupby(['weekday', 'hour'])['total_demand']
        .mean()
    )

    # 3. In 7×24‑Matrix (168 Werte) bringen, fehlende Kombis als 0
    grid = np.zeros((7, 24))
    for (wd, hr), val in weekly.items():
        if 0 <= wd <= 6 and 0 <= hr <= 23:
            grid[wd, hr] = val

    # 4. Flache Form 168‑Dim‑Vektor
    profile_168 = grid.flatten()  # 7*24 = 168

    # 5. normalisieren auf Maximum (Max = 1)
    max_val = profile_168.max()
    if max_val > 0:
        normalized = profile_168 / max_val
    else:
        normalized = profile_168.copy()

    weekly_profiles[city] = normalized


# ===================================================================
# 4. Kosinus‑Ähnlichkeit zu Berlin (Berlin als Source City)
# ===================================================================

# Stelle sicher, dass Berlin in weekly_profiles ist
CITY_SOURCE = 'Berlin'

if CITY_SOURCE not in weekly_profiles:
    raise ValueError(f"Stadt '{CITY_SOURCE}' ist nicht in weekly_profiles.")

X = np.vstack(list(weekly_profiles.values()))   # 7×168 Matrix
cities = list(weekly_profiles.keys())

# Index von Berlin bestimmen
berlin_idx = cities.index(CITY_SOURCE)

# Kosinus‑Ähnlichkeit zu Berlin für jede Stadt
similarity_to_berlin = cosine_similarity(
    X,
    X[berlin_idx:berlin_idx+1]  # 1×168, Berlin als Quelle
).flatten()

# Dictionary: Stadt → Kosinus‑Ähnlichkeit zu Berlin
sim_dict = dict(zip(cities, similarity_to_berlin))

print("\n" + "="*70)
print("Kosinus‑Ähnlichkeit zu Berlin (Source City)")
print("="*70)
for city in sim_dict:
    print(f"{city:30s}: {sim_dict[city]:.4f}")

# Optional: als CSV speichern
df_out = pd.DataFrame({
    'city': list(weekly_profiles.keys()),
    'cosine_similarity_to_berlin': list(sim_dict.values())
})
df_out.to_csv(OUTPUT_CSV, index=False)
```
